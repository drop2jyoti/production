# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Load the libraries as required.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import ElasticNet


import warnings

warnings.filterwarnings('ignore')
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fires_dt.info()
```

```python
fires_dt.head(4)
```

# Get X and Y

Create the features data frame and target data.

```python
X = fires_dt.iloc[:, :-1]
Y = fires_dt.iloc[:, -1:]
#print(X, Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)


print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


```python
num_features = ['coord_x', 'coord_y', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
num_trans_features = ['ffmc', 'dmc', 'dc']
cat_features = ['month', 'day']
```

### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
# preproc1
preproc1 = ColumnTransformer([
    ('robust_scaler',RobustScaler(),num_features),
    ('onehot', OneHotEncoder(handle_unknown='ignore'), cat_features)
])

preproc1
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
# preproc2
preproc2 = ColumnTransformer([
    ('standard_scaler', StandardScaler(), num_features),
    ('transform', PowerTransformer(method='yeo-johnson'), num_trans_features),
    ('onehot', OneHotEncoder(handle_unknown='ignore'), cat_features)
])

preproc2
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
#scoring = ['neg_mean_absolute_error', 'neg_mean_squared_error', 'r2', 'explained_variance']
scoring = [ 'neg_mean_squared_error']
```

```python
base_regressor = LinearRegression()
```

```python
# Pipeline A = preproc1 + baseline
pipeline_A = Pipeline([
    ('preproc', preproc1),
    ('regressor', base_regressor)
])
pipeline_A
```

```python
# Pipeline B = preproc2 + baseline
pipeline_B = Pipeline([
    ('preproc', preproc2),
    ('regressor', base_regressor)
])
pipeline_B
```

```python
# alternative_regressor = LogisticRegression()
alternative_regressor = RandomForestRegressor()
```

```python
# Pipeline C = preproc1 + advanced model

pipeline_C = Pipeline([
    ('preproc', preproc1),
    ('regressor', alternative_regressor)
])
pipeline_C
```

```python
# Pipeline D = preproc2 + advanced model
pipeline_D = Pipeline([
    ('preproc', preproc2),
    ('regressor', alternative_regressor)
])
pipeline_D
    
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
param_grid_A = {
    'regressor__fit_intercept': [False, True],
    'regressor__positive': [False, True],
    'regressor__copy_X': [True, False],
}
```

```python

grid_search_1 = GridSearchCV(estimator=pipeline_A, param_grid=param_grid_A, cv=5, scoring=scoring, return_train_score=True, refit = 'neg_mean_squared_error')
gs1 = grid_search_1.fit(X_train, Y_train)

```

```python
print(f"### Pipeline A\n**Best score:** {gs1.best_score_}  \n**Best params:** {gs1.best_params_}")
```

```python
grid_search_2 = GridSearchCV(estimator=pipeline_B, param_grid=param_grid_A, cv=5, scoring=scoring, return_train_score=True, refit='neg_mean_squared_error')
gs2 = grid_search_2.fit(X_train, Y_train)
print(f"### Pipeline B\n**Best score:** {gs2.best_score_}  \n**Best params:** {gs2.best_params_}")
```

```python
param_grid_B = {
    'regressor__bootstrap': [True, False],
    'regressor__max_depth': [2, 4, 6],
}
```

```python

grid_search_3 = GridSearchCV(estimator=pipeline_C, param_grid=param_grid_B, cv=5,scoring=scoring, refit='neg_mean_squared_error', verbose=False)
gs3 = grid_search_3.fit(X_train, Y_train)
print(f"### Pipeline C\n**Best score:** {gs3.best_score_}  \n**Best params:** {gs3.best_params_}")
```

```python
grid_search_4 = GridSearchCV(estimator=pipeline_D, param_grid=param_grid_B, cv=5, scoring=scoring,refit='neg_mean_squared_error', verbose=False)
gs4 = grid_search_4.fit(X_train, Y_train)
print(f"### Pipeline D\n**Best score:** {gs4.best_score_}  \n**Best params:** {gs4.best_params_}")
```
